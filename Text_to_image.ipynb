```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import random

# โหลดโมเดลที่ฝึกมาแล้ว
model_id = "runwayml/stable-diffusion-v1-5"  # เปลี่ยนเป็น model ที่คุณโหลดไว้ถ้าไม่ได้ต่อเน็ต
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")

# รับข้อความอธิบายจากผู้ใช้
user_prompt = input("📝 Describe the picture you want to generate (English): ")

# เพิ่มรายละเอียดให้ชัดเจนขึ้น
detailed_prompt = f"{user_prompt}, ultra-detailed, high-resolution, cinematic lighting, intricate textures, beautifully rendered"

# ตั้งค่าการสุ่ม เพื่อให้ภาพหลากหลาย หรือกำหนด seed ถ้าอยากได้ผลลัพธ์เดิม
generator = torch.Generator("cuda").manual_seed(random.randint(0, 999999))

# สร้างภาพด้วย prompt และการตั้งค่าที่ปรับปรุง
image = pipe(
    detailed_prompt,
    num_inference_steps=75,       # เพิ่มจำนวนขั้นตอนให้รายละเอียดดีขึ้น
    guidance_scale=10.0,          # เพิ่มการเน้นตาม prompt
    height=768,                   # ขนาดภาพ
    width=768,
    generator=generator
).images[0]

# บันทึกและแสดงภาพ
image.save("high_quality_image.png")
display(image)

print("\n✅ Image generated successfully in high quality!")
```
